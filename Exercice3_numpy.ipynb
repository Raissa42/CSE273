```python
#Create a numpy array electricity using np.random.randint
electricity = np.random.randint(120, 650, size=(4, 7))
print(electricity)
#Value between 120 and 650
#Shape 4,7
#Task 2: print the following attributes of the array:shape,number of dimension,total number of element,and data type
print(electricity.shape)
print(electricity.ndim)
print(electricity.size)
print(electricity.dtype)
```

```python
#Task B1:Hostel-wise Metrics 
#Find the average electricity usage per hostel,eh maximum usage per hostel,and identify which hostel has the highest average usage
average_per_hostel = np.mean(electricity, axis=0)
print(np.round(average_per_hostel,2))
print(np.max(average_per_hostel))
#Task B2:Day-wise metrics
#Calculate the total electicity consumption per day and identify the day with the highest total campus consumption(must choose the correct axis value for each operation)
total = (np.sum(electricity,axis=0))
print(total)
print("Highest Total campus consumption:")
print(np.max(total))
```
