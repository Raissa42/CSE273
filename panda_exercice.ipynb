```python
import numpy as np
import pandas as pd
```

```python
Data ={
    "Name" : ["Aman","Saurav","Karan","Shubham","Abhishek",None,"Sameer"],
    "Age" : [22,23,np.nan,22,25,24,26],
    "Marks":[85,np.nan,88,90,56,72,np.nan],
    "Subject" : ["Math","Physics","Math","Chemistry","Math","Physics","Chemistry"]
}
```

```python
#df = panda dataframe(use np.nan for missing value)
df = pd.DataFrame(Data)
print(df)
#check shape
print(df.shape)
#display the datatype of the column
print(df.columns)
#generate a statistical summary(mean/min/max)for the numerical columns
df.describe()
```

```python
#Filter and group the data
#Selection:create a view showing only the 'Name' and 'Marks' column
view = df[["Name","Marks"]]
print(view)
#Filtering: Select student who have scored more than 80marks
print("Student scoring more than 80 marks: ")
print(df[df["Marks"] > 80])
#complex filtering: select student older than 22 who also scored more than 80
print("MORE THAN 22 AND GREATER THAN 80 : ")
print(df[(df["Age"] > 22) & (df["Marks"] > 80)])
#Aggregation: count how many students are enrolled in each subject
print("Number of student enrolled in each course: ")
print(df["Subject"].value_counts())
```

```python
#Order the data to find rankers
#Sort the dataframe by 'Marks'in descending order
print(df.sort_values("Marks",ascending=False))
#Use a specific Pandas function to find the top 2 highest scores(Scholarship Candidates) without sorting the whole dataframe manually
top2 = df.nlargest(2,"Marks")
print("TOP 2 HIGHEST")
print(top2)
#Identify the student with the lowest marks(remedial candidate)
print("Lowest : ")
lowest = df.nsmallest(1,"Marks")
print(lowest)
```
