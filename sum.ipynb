```python
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
```

```python
print(np.sum(arr))
```

```python
#Sum element of each column
print(np.sum(arr,axis=0))
```

```python
#Sum of each row
print(np.sum(arr,axis=1))
```

```python
#Where in numpy
ar = np.array([10,-5,20,-3,15])
#If and else(condition,true,false)
result = np.where(ar < 0,0 ,ar)
print(result)
```

```python
arr = np.array([1,2,3,4,5])
result = np.where(arr % 2 !=0, -1 , arr)
print(result)
```

```python
marks = np.array([45,78,32,90,55])
status = np.where(marks >= 50,1,0)
print(status)
```

```python
#finding mean value
data = np.array([10,20,30,40,50])
mean_val = np.mean(data)
result = np.where(data > mean_val,mean_val,data)
print(result)
```

```python
#Applying where in 2d array
mat = np.array([[10,60,30],[80,20,90]])
result = np.where(mat >= 50,1,0)
print(result)
```
