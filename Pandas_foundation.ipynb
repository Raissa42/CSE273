# GroupBy in Pandas

```python
data ={
    "Name" : ["Aman","Saurav","Karan","Shubham","Abhishek",None,"Sameer"],
    "Age" : [22,23,np.nan,22,25,24,26],
    "Marks":[85,np.nan,88,90,56,72,np.nan],
    "Subject" : ["Math","Physics","Math","Chemistry","Math","Physics","Chemistry"]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Grupby basic/mean marks per subject
df.groupby('Subject')['Marks'].mean()
```

```python
df.groupby('Subject')['Name'].count()#Count student per subject /Ignore missing value
```

```python
df.groupby('Subject').size()# will include the count value in the count
```

```python
#Multiple aggregations
df.groupby('Subject')['Marks'].agg(['mean','max','min'])
```

```python
#Groupby multiple columns
df.groupby(['Subject','Age'])['Marks'].mean() # ->(in which column to apply)[what to apply]
```
