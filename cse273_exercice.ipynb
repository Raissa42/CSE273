```python
#You r managing the electronics section of a store.you need to track the sales of 3 products:Headphones,smartwatches,and speakers over 4 days(monday to thursday)
#row 0:headphone
#row 1:smartwatches
#row 2:speakers
```

```python
import numpy as np
dily_sales = np.array([[1500],[3000],[5000]])
print(dily_sales.shape)
print(dily_sales.ndim)
```

```python
sold = np.array([[10,15,8,12],[5,7,4,6],[2,3,2,4]])
print(sold.shape)
print(sold.ndim)
print(sold)
```

```python
print("Adding bonus: ")
print(sold + 2)
```

```python
print(dily_sales * sold)
#print(np.dot(sold,dily_sales))
```

```python
#Access the sale data for Smartwatches(row1) for all 4 days#slicing and indexing
#Acess the sales data for all product for the first 2 days
#Access sales of speakers on thursday using double indexing
print(sold[0])
print("All product of the first 2 days: ")
print(sold[:, :2])
print("Sales of speakers on thrusday:")
print(sold[:, 3:])

```

```python
#RESHAPING AND LINESPACE
#Flatten the 3*4daily sales array into 1D array of 12 element using .reshape()
print("Reshapen sales array:")
print(sold.reshape(-1))
#Convert the sales data into 2*6 matrix
print("2 * 6:")
print(sold.reshape(2,6))
#Using np.linspace to create an array of 10 equal inervall between 0 and 100 to represent "Discount percentage"
array = np.linspace(1,100,10)
print("Discount Percentage:")
print(array)
```

```python
#The random module
#Use np.random.randint to generate a new 3*4 array of "Predicted Sales "for the next week with values ranging from 0 to 20.
predicted_sales = np.random.randint(0,20,size=(3,4))
print("Predicted sales from 0 t0 20")
print(predicted_sales)
#Generate a random array of 5 numbers between 0 and 1 using np.random.randint
random = np.random.randint(0,1,5)
print("Random from 0 to 1: ")
print(random)
#Use np.sum() to find:total items sold over all 4 days,total items sold for each product(Sum across rows)
print("Total sold over 4 days: ")
print(np.sum(sold,axis=0))
```
